Subplots with pyplot:

plt.subplots returns 2 objects: a container for all plots (figure) and the positiion of each plot (axes)
* figure acts as a container for all plots and has methods for customizing the appearance and behavior for the plots in the container
* 2 ways of creating subplots:
    * f = plt.figure(), ax1 = f.add_subplot(nrows, ncols, plot_number)
    * f, ax = plt.subplots(nrows, ncols)

```python
# subplot example
import matplotlib.pyplot as plt

# 1st method
f = plt.figure()
ax1 = f.add_subplot(2, 1, 1)
ax2 = f.add_subplot(2, 1, 2)
ax1.plot([1, 2, 3, 4])
ax2.plot([-1, -2, -3, -4])
```

```python
# 2nd method
f, ax = plt.subplots(2)
ax[0].plot([1, 2, 3, 4])
ax[1].plot([-1, -2, -3, -4])
```

### Object-Oriented Programming

class - can contain own objects, methods, and attributes

methods - specific functions associated with the class
* can be called with objectname.method()

attributes - values that are stored inside the class object when created

```python
class MyClass:
    # defining the class to take inputs
    def __init__(self, lst):
        # creating an attribute of the class
        self.data = lst
        
    # defining method for the class
    def show(self):
        print(self.data)
        
```

```python
my_list = MyClass([1, 2, 3, 4, 5])
my_list.show()
```
