# Decision Tree

```python
from sklearn import tree
X = [[0, 0], [1, 1]]
y = [0, 1]
clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
```

```python
clf.predict([[2., 2.]])
# 2, 2 gives prediction of 1
```

```python
clf.predict_proba([[2., 2.]])
# predict_proba returns the probability of each value (for classifier only)
```

```python
tree.plot_tree(clf.fit(X, y))
# plotting the decision tree
```

# Practice with Pima Diabetes Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df.shape
```

```python
df.head()
```

## Train Test Split

```python
# split the dataset into train and test by cutting off the dataset at 80% of index
train = df.iloc[0:int(df.shape[0] * 0.8)].copy()
train.shape
```

```python
test = df.iloc[int(df.shape[0] * 0.8):].copy()
test.shape
```

## Extracting Feature Names

```python
cols = train.columns[:-1].tolist()
cols
```

## Target Variable name

```python
target = train.columns[-1]
target
```

## Retrieving X and y

```python
X_train = train[cols]
X_test = test[cols]
X_train.head()
```

```python
y_train = train[target]
y_test = test[target]
y_train
```

# Modelling

```python
# import Decision Tree model
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model
```

```python
# train model with train set
model.fit(X_train, y_train)
```

```python
# predict based on test set
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize = (20 ,10))
tree = plot_tree(model, feature_names = cols, filled = True, fontsize = 10)
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_tree = export_graphviz(model, feature_names = cols, filled = True)
graphviz.Source(dot_tree)
```

```python
# feature importance
model.feature_importances_
```

```python
sns.barplot(x = model.feature_importances_, y = cols)
```

```python
# Accuracy

1 - (abs(y_test - y_pred).sum())/len(y_test)
```

```python
# Accuracy with sklearn function
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
model.score(X_test, y_test)
# score takes the test set as a whole and returns the score, but this can only be used when there is y_test
```
